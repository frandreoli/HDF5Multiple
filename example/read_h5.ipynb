```python
import h5py
import numpy as np
import os
print("The working directory is: ", str(os.getcwd()))
file_name = 'test.h5'
```

```python
#Prints the entire structure of an H5 file.
def print_structure(name, obj):
    # Indent based on the depth of the object in the hierarchy
    depth = name.count('/')
    to_add = "\u00A6\n"
    if depth>0:
        to_add = ""
        name = name[name.rfind("/")+1:]
        if depth>1:
            indent = ("\u00A6   "*(depth-1))+ "\u251c-> "
        else:
            indent = "\u251c-> "
    else:
        indent = ""
    if isinstance(obj, h5py.Group):
        description = [f"{key}: {value}" for key, value in obj.attrs.items()]
        description = ",".join(description)
        print(f"{to_add}{indent}{name} ({description})")
    else:
        print(f"{to_add}{indent}{name}: {obj[()]}")

def explore_h5_file(file_path):
    with h5py.File(file_path, 'r') as f:
        print(f"Exploring structure of: {file_path}\n" + "="*30)
        f.visititems(print_structure)
```

```python
explore_h5_file(file_name);
```
